Verificar que el desvio estandar sea igual a $\sqrt{ pq }$ de la variable dummie.

````python
mfr_type_dummies.mean()
````

````python
mfr_type_dummies.std()
````

````python
def calc_sqrt_p_q(s):
    # p = sum/count
    # q = 1 - p
    p = s.sum() / s.count()
    q = 1 - p
    return np.sqrt(p*q)
    
mfr_type_dummies.apply(calc_sqrt_p_q)
````

#### Elegir un fabricante y resolver las siguientes propuestas:
       
   * Si en un supermercado entran 2000 personas de las cuales 50 van a comprar cereales y elijen una caja aleatoriamente por persona:
       
       * Describir la función binomial (_n, p_), que modela el problema y graficar un histograma obteniendo muestras que sigan esa tendencia con la función ``` np.random.binomial()```. (Ver notebook de clase 9)

````python
mfr_counts = cereal['mfr'].value_counts()
mfr_p = mfr_counts.apply(lambda x: x / mfr_counts.sum())
mfr_p
````

````python
# definimos la cantidad de ensayos.
cantidad_ensayos = 2000

# definimos la cantidad de lanzamientos por ensayo.
lanzamientos = 50

# definimos la probabilidad de obtener cara en cada lanzamiento.
probabilidad_exito = mfr_p[6]

# binomial devuelve el número de éxitos de cada vez que se realizo un ensayo de 8 lanzamientos.
samples_binomial = np.random.binomial(n = lanzamientos, p = probabilidad_exito, size = cantidad_ensayos)

# construímos un gráfico.
eje_x = 'Número de compras por ensayo de 50 clientes'
sns.distplot(samples_binomial, axlabel = eje_x, kde = 0);
````
