### 21. A partir de las variables seleccionados, ¿observas una mayor incidencia de alguna de las variables y la performance del estado en alguno de los scores? ¿Consideras importante inferir alguna conclusión para orientar a los directivos de la universidad y/o profundizar el análisis?  

```python


satWithGDP = pd.merge(satScore, GDP, on='State', how='left')
medianGDP = satWithGDP['GDP2016'].median()
satWithGDP['GDP2016'] = satWithGDP['GDP2016'].fillna(medianGDP)
satWithGDP.info()

#Hace un left join del dataframe original con el GDP
#Reemplaza los valores faltantes del GDP por la mediana
```

```python
GINI = pd.read_csv('GINI2015_martin.csv')

satWithGINI = pd.merge(satWithGDP, GINI, on='State', how='left')
medianGINI = satWithGINI['Gini'].median()
satWithGINI['Gini'] = satWithGINI['Gini'].fillna(medianGINI)
satWithGINI.info()
```

```python
PopAndDesity = pd.read_csv('PopulationAndDensity_martin.csv')

satWithPop = pd.merge(satWithGINI, PopAndDesity, on='State', how='left')
medianPop = satWithPop['Population'].median()
medianDensity = satWithPop['Density'].median()
satWithPop['Population'] = satWithPop['Population'].fillna(medianPop)
satWithPop['Density'] = satWithPop['Density'].fillna(medianDensity)

satWithPop.head()
```

```python
Immigrants = pd.read_csv('Immigrants_martin.csv')
satWithImmigrants = pd.merge(satWithPop, Immigrants, on='State', how='left')
medianImmigrants = satWithImmigrants['Immigrants'].median()
satWithImmigrants['Immigrants'] = satWithImmigrants['Immigrants'].fillna(medianImmigrants)


satWithImmigrants.head()
```

### 22. Graficá algunos scatterplots. Usá algún gráfico de Seaborn para presentar scatterplots múltiples
