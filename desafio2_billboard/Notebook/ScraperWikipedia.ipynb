```python
df = pd.read_csv('billboard.csv',encoding='latin-1')
```

```python
def getGenre(artist_name):
    artist_name = artist_name.replace (" ","%20")
    print("http://www.allmusic.com/search/all/{}".format(artist_name))
    result = requests.get("http://www.allmusic.com/search/all/{}".format(artist_name),headers=headers)
    c = result.content
    soup = BeautifulSoup(c,'html.parser')
    ulSearch = soup.findAll('ul', attrs={'class': 'search-results'})
    liartist = ulSearch[0].findAll('li', attrs={'class': 'artist'})
    aux = [x.findAll('div', attrs={'class':'genres'})[0] for x in liartist if len(x.findAll('div', attrs={'class':'genres'})) > 0]
    if len(aux) > 0:
        print(aux[0])
#        return aux[0].findAll('div', attrs={'class':'genres'})[0].next
        return aux[0]
    else:
        return np.nan
```

```python
result = [(x,getGenre(x)) for x in df['artist.inverted'].values]
```
