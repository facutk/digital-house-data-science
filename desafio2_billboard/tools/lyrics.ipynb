```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
def arrange_artist(artist):
    return ' '.join(artist.split(', ')[::-1])

def norm(string):
    return re.sub(r'\W+', '-', string.lower())

def gen_url(artist, track):
    base_url = 'http://www.songlyrics.com'
    return '%s/%s/%s-lyrics/'%(base_url, norm(arrange_artist(artist)), norm(track))

def lyrics(artist, track):
    url = gen_url(artist, track)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    l = soup.find("p", {"id": "songLyricsDiv"}).text
    if l.startswith("Sorry, we have no"):
        return None
    return l
```

```python
lyrics('Lady Gaga', 'telephone')
```

```python
import pandas as pd
df = pd.read_csv('../Data/billboard.csv', encoding='latin1')

def process(df):
    print(df['artist.inverted'])
    df['lyrics'] = lyrics(df['artist.inverted'], df['track'])
    return df

artist_track = df[['artist.inverted', 'track']]
artist_track = artist_track.apply(process, axis = 1)

artist_track.to_csv('./lyrics.csv', encoding='utf-8', index = False)
```
